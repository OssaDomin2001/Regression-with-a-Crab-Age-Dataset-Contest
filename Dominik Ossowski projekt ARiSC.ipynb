```python
lb=LabelEncoder()
df_train['Sex'] = lb.fit_transform(df_train['Sex'])
df_train
```

```python
df_train = df_train[(df_train[df_train.columns[df_train.columns != 'Sex']] != 0).all(axis=1)].copy()
```

```python
lb=LabelEncoder()
df_test['Sex'] = lb.fit_transform(df_test['Sex'])
df_test
```

```python
merged_df = pd.concat([df_train, df_test], axis=0)
merged_df
```

Podczas przeglądania prac natknąłem się ciekawe rozwiązanie w zakresie feature engineering dla podanego data setu - polega ono na dodaniu kolumn reprezentujących różne stosunki, np.: długości do średnicy czy wagi muszli do wagi całego kraba.

```python
merged_df['Total_weight'] = merged_df[['Shucked Weight', 'Viscera Weight', 'Shell Weight']].sum(axis=1)
merged_df
```

```python
merged_df["Length_to_Diameter"] = merged_df["Length"] / merged_df["Diameter"]
merged_df["Height_to_Diameter"] = merged_df["Height"] / merged_df["Diameter"]
merged_df["Shucked_to_Weight"] = merged_df["Shucked Weight"] / merged_df["Weight"]
merged_df["Viscera_to_Weight"] = merged_df["Viscera Weight"] / merged_df["Weight"]
merged_df['Shell_to_Weight'] = merged_df['Shell Weight'] / merged_df['Weight']
merged_df["Volume"] = math.pi * (merged_df["Diameter"]*merged_df['Length']*merged_df["Height"])/3
merged_df
```

```python
df_merged_train = merged_df[merged_df['Age'].notnull()]
df_merged_train
```

```python
df_merged_test = merged_df[merged_df['Age'].isnull()==True]
df_merged_test
```

```python
X = df_merged_train.drop('Age', axis=1)
y = df_merged_train['Age']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
print(X_train.shape)
print(X_test.shape)
```

```python
scaler = MinMaxScaler()

X_train_normalized = scaler.fit_transform(X_train)
X_test_normalized = scaler.transform(X_test)
```

Wykorzystałem algorytm XGBoost do rozwiązania problemu

```python
dtrain = xgb.DMatrix(X_train_normalized, label=y_train)


params = {
    'learning_rate': 0.4,
    'n_estimators': 10,
    'max_depth': 4,
    'subsample': 0.8,
    'colsample_bytree': 0.8,
    'reg_alpha': 0.5,
    'reg_lambda': 0.5,
    'gamma': 0.0,
    'min_child_weight': 1,
    'objective': 'reg:squarederror'
}
```

```python
xgb_model = xgb.train(params, dtrain)

dtest = xgb.DMatrix(X_test_normalized)

y_pred = xgb_model.predict(dtest)

mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)

print(f"Root Mean Squared Error (RMSE): {rmse}")
```

```python
results_df_rf = pd.DataFrame({'Actual':y_test,'Predicted':y_pred})
results_df_rf
```
